```python
x = np.arange(1,11)
y=np.random.randint(10,100,size= 10)
data=np.random.randn(100)
print(x)
print(y)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x, y, '--', marker='o',color='black')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot (Dashed Line)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(x, y, marker='*',color='red')
plt.title("Scatter Plot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.bar(x, y,color='red')
plt.title("Bar Plot")
plt.grid(True)
plt.show()
```
